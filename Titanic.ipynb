```python
import pandas as pd
df = pd.read_csv("titanic.csv")
df
```

<h2>We want to drop the name, sibsp, parch, passangerid, cabin, ticker and embarked columns</h2>

```python
inputs = df.drop(columns=["PassengerId", "Name", "SibSp", "Parch", "Ticket", "Cabin", "Embarked", "Survived"], axis="columns")
target = df["Survived"]
```

<h2>Its time to train our data </h2>

```python
from sklearn.preprocessing import LabelEncoder
```

<h2>Indicates the label encoders</h2>

```python
le_pclass = LabelEncoder()
le_sex = LabelEncoder()
le_age = LabelEncoder()
le_fare = LabelEncoder()
```

<h1>Testing inputs on the Passanger class to calculate survival rate</h1>

```python
inputs["p_class"] = le_pclass.fit_transform(inputs["Pclass"])
inputs["sex"] = le_pclass.fit_transform(inputs["Sex"])
inputs["age"] = le_pclass.fit_transform(inputs["Age"])
inputs['fare'] = le_pclass.fit_transform(inputs["Fare"])
inputs["age"].fillna(inputs["age"].mean(), inplace=True)
```

```python
inputs.head()
```

```python
inputs_pclass = inputs.drop(["Pclass", "Age", "Sex", "Fare"], axis="columns")
inputs_pclass
```

```python
from sklearn.model_selection import train_test_split
import seaborn as sn
import pandas as pd
X_train, X_test, y_train, y_test = train_test_split(inputs_pclass, target)
```

```python
X_train.head()
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
