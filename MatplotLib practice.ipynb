```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
days = [1, 2, 3, 4, 5]
max_temp = [100, 89, 92, 93, 94]
min_temp = [60, 65, 62, 71, 70]
avg_temp = [85, 80, 87, 90, 79]
```

```python
plt.xlabel("Days")
plt.ylabel("Temparatures")
plt.title("Weather of a hot city")
plt.plot(days, max_temp, label="Max temp")
plt.plot(days, min_temp, label="Min temp")
plt.plot(days, avg_temp, label="Avg temp")
plt.legend(loc="best", shadow=True, fontsize="small")
```

```python
company = ["GOOG", 'AMZN', "MSFT", "FB"]
money = [100, 89, 120, 140]
profit=[180, 169, 200, 220]
ypos = np.arange(len(company))
ypos
```

```python
plt.xticks(ypos, company)
plt.ylabel("revenue")
plt.title("Tech stocks")
plt.bar(ypos, profit, label="Profit")
plt.bar(ypos, money, label="Revenue")
plt.legend()
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel("Sugar range")
plt.ylabel("No of patients")
plt.hist([blood_sugar_men, blood_sugar_women], bins= [80, 100, 125, 150], rwidth=0.95, color=["black", "y"], label=["men", "women"])
plt.legend()
```
